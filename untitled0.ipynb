```python
import pandas as pd

url = "https://raw.githubusercontent.com/anagovbr/hidro-dados-estacoes-convencionais/main/Inventario_Estacoes_Hidrologicas_04-08-2023.csv"

df = pd.read_csv(url, sep=';', encoding='latin1')  # o arquivo usa ; e encoding latin1

print(df.head())

# 2. Renomear colunas para facilitar manipulação
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

# 3. Mostrar as primeiras linhas e info básica
print("Primeiras linhas:")
print(df.head())

print("\nInformações gerais:")
print(df.info())

# 4. Verificar quantos valores faltantes por coluna
print("\nValores faltantes por coluna:")
print(df.isnull().sum())

# 5. Limpar: remover linhas com qualquer dado faltante
df_limpo = df.dropna()

print(f"\nApós limpeza, número de linhas: {len(df_limpo)}")

# 6. Resetar índice após limpeza
df_limpo = df_limpo.reset_index(drop=True)

# 7. Salvar CSV limpo localmente
df_limpo.to_csv('inventario_estacoes_limpo.csv', index=False)

print("\nArquivo CSV limpo salvo como 'inventario_estacoes_limpo.csv'")
```
